Example:
 to request search results for the search term "memory cards" sorted by "high to low price",

```python
params = {
    'api_key'       : api_key,
    'type': 'search',
    'amazon_domain': 'amazon.com',
    'search_term': 'apple iphone 15 pro max',
    'direct_search':'true'
    }

# make the http GET request to Rainforest API
api_result = requests.get('https://api.rainforestapi.com/request', params)
```

## Show json Pretty

```python
json_formatted_str = json.dumps(api_result.json(), indent=2)

print(json_formatted_str)
```

```python
result=api_result.json()
type(result)
```

```python
product = result['search_results']
print(len(product))
print(type(product))
```

```python
product_df = pd.DataFrame()
for i in range(0,len(product)-1):
    temp_df =  pd.DataFrame.from_dict( product[i], orient="index").T
    product_df = pd.concat([product_df , temp_df])
    # temp_df = pd.DataFrame({'date' : product[i].keys() , 'date_value' : product[i].values() }).T
    
```
